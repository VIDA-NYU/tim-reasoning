```python
import json
import random
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from tim_reasoning import SessionManager


idx_traversed = {"pinwheels": 0, "quesadilla": 0, "oatmeal": 0}
# idx_traversed = {"pinwheels": 0, "quesadilla": 0}
all_data = []
total_rows = 0
for k in idx_traversed.keys():
    with open(f"examples/test/{k}_perception_outputs.json", "r") as f:
        data = f.read()
    data = json.loads(data)
    total_rows += len(data)
    all_data.append(data)

total_recipe = len(all_data)
key_map = list(idx_traversed.keys())

result = []
while True:
    if len(result) >= total_rows:
        break
    k_idx = random.randint(0, total_recipe-1)
    recipe = key_map[k_idx]
    curr_row = idx_traversed[recipe]
    if len(all_data[k_idx]) > curr_row:
        result.append(all_data[k_idx][curr_row])
        idx_traversed[recipe] += 1

# run sess manager

sm = SessionManager(patience=1)
output = []
for row in result:
    output.append(sm.handle_message(message=[row]))
```

```python
correct = 0
num_class_pred = []
confusion_matrix = np.zeros((total_recipe, total_recipe))

for i in range(len(result)):
    result_task_id = result[i]['session']['task_id']
    
    match = False
    for j in range(len(output[i])):
        num_class_pred.append(len(output[i]))
        if output[i][j]:
            if result_task_id == output[i][j]['task_id']:
                match = True
                break
    
    if match:
        correct += 1
        true_class = key_map.index(result_task_id)
        pred_class = key_map.index(result_task_id) 
        confusion_matrix[true_class, pred_class] += 1
    else:
        print(f"i = {i}, result_task_id = {result_task_id}, \n result[i] = {result[i]}, \n output[i] = {output[i]} \n\n")

task_rec_accuracy = correct / len(result)
```

```python
sns.heatmap(confusion_matrix, annot=True, cbar=False, xticklabels=key_map, yticklabels=key_map)

plt.xlabel("Predicted")
plt.ylabel("True")
plt.title("Confusion Matrix")

plt.show()
```

```python
correct_map = {'pinwheels': {}, 'quesadilla': {}, 'oatmeal': {}}
total_map = {'pinwheels': {}, 'quesadilla': {}, 'oatmeal': {}}
correct = 0 
for i in range(len(result)):
    result_task_id = result[i]['session']['task_id']
    result_step_id = result[i]['session']['step_id']
  
    for j in range(len(output[i])):
        if result_task_id == output[i][j]['task_id'] and result_step_id == output[i][j]['step_id']:
            correct += 1
            if str(result_step_id) in correct_map[result_task_id]:
                correct_map[result_task_id][str(result_step_id)] += 1
            else:
                correct_map[result_task_id][str(result_step_id)] = 1
            break
    if str(result_step_id) in total_map[result_task_id]:
        total_map[result_task_id][str(result_step_id)] += 1
    else:
        total_map[result_task_id][str(result_step_id)] = 1

step_accuracy = correct / len(result)

print("Step accuracy:", step_accuracy)
```

```python
fig, axes = plt.subplots(nrows=1, ncols=3, figsize=(10, 5))

for i, (recipe, axis) in enumerate(zip(correct_map, axes)):
    steps = list(correct_map[recipe].keys())
    accuracy = [correct_map[recipe][step] / total_map[recipe][step] for step in steps]
  
    axis.plot(steps, accuracy, marker='o')
    axis.set_xlabel('Steps')
    axis.set_ylabel('Accuracy')
    axis.set_title(recipe)

fig.tight_layout()
plt.show()
```
